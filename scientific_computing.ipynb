c) Calculating the Volume Under a Surface



```python
def calculate_volume(n):
    """
    Calculate the volume under the surface z = x^2 + y^2
    over the region 0 ≤ x, y ≤ 1 using midpoint rule.

    Args:
    n (int): Number of intervals along each axis.

    Returns:
    float: Approximated volume.
    """
    dx = 1.0 / n
    dy = 1.0 / n
    volume = 0.0

    for i in range(n):
        for j in range(n):
            x = (i + 0.5) * dx
            y = (j + 0.5) * dy
            z = x**2 + y**2
            volume += z * dx * dy

    return volume

try:
    n = int(input("Enter the number of intervals (e.g., 100): "))
    if n <= 0:
        raise ValueError("Number must be positive.")
    volume = calculate_volume(n)
    print(f"Approximated volume under z = x² + y² over [0,1]x[0,1] is: {volume:.5f}")
except ValueError as e:
    print("Invalid input:", e)
```

d) Compiled vs. Interpreted Programming Languages and Python



Compiled languages  the source code is  translated into machine code before executon using  a complier

interpterted languages the source code is executed by an interprator and no seperate  compiation is required.

execution is faster in compiled languages compared to interprated languages because the code is already translated into low level languages.




Python is an interpreted language. When you run a Python script, the Python interpreter reads and executes the code line by line.

Python first compiles the source code into bytecode (files with a .pyc extension).
This bytecode is then interpreted by the Python Virtual Machine (PVM).
